```python
# import libary

import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
```

```python
# menampilkan Gambar Asli

uang = cv2.imread('100k.jpg')
plt.imshow(cv2.cvtColor(uang, cv2.COLOR_BGR2RGB))
cv2.imwrite('100k.jpg', uang)
```

```python
# menampilkan gambar ke grayscale

uang_gray = cv2.imread('100k.jpg')
plt.imshow(cv2.cvtColor(uang_gray, cv2.IMREAD_GRAYSCALE))
cv2.imwrite('100k.jpg', uang)
```

```python
# mengubah citra ke biner

uang_gray = cv2.imread('100k.jpg')
ret,biner = cv2.threshold(uang_gray,150,255,cv2.THRESH_BINARY)
cv2.imwrite('100k.jpg', biner)
```

```python
# menampilkan original image  dan satunya di beri canny

uang_tresh_hi = cv2.imread('100k.jpg',0)
edges = cv2.Canny(uang,100,200)
 
plt.subplot(121),plt.imshow(uang,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Canny Image'), plt.xticks([]), plt.yticks([])
 
plt.show()
```

```python
# menampilkan 1 gambar dengan metode canny

uang_canny = cv2.Canny(biner, threshold1=90, threshold2=110)
plt.imshow(cv2.cvtColor(uang_canny, cv2.COLOR_GRAY2RGB))
cv2.imwrite('100k.jpg', uang_canny)
```

```python
# menkonversi citra ke nilai array

uang_canny[:,1]
```

```python
# menampilkan citra ke dalam bentuk diagram 

histogram1=uang_canny[...,0].flatten()
histogram_uang1 = plt.hist(histogram1,256)
plt.show()
```

```python
# menampilkan array dari histogram di atas

histogram_uang1[1]
```

```python
# menampilkan citra ke dalam bentuk diagram dan diberi batasan 256

histogram2=uang_canny[...,1].flatten()
histogram_uang2 = plt.hist(histogram2,256)
plt.show()
```

```python
# menampilkan array dari histogram di atas

histogram_uang2[1]
```

```python
watermark_no = 0
watermark_yes = 0
for index,value in enumerate(histogram_uang1[1]):
    if (value >= 0 and value <= 127.5) and (histogram_uang2[1][index] >= 0 and histogram_uang2[1][index] <= 127):
        watermark_no += 1
    
    if (value >= 128 and value <= 256) and (histogram_uang2[1][index] >= 128 and histogram_uang2[1][index] <= 256):
        watermark_yes += 1
watermark_no = watermark_no/histogram_uang1[1].shape[0]*100
print(watermark_no)
print(watermark_yes)
```

```python
watermark_tidak_ada = False
watermark_terdeteksi = False

if  watermark_tidak_ada>127.5:
    watermark_tidak_ada = True
    
if  watermark_no <255:
    watermark_terdeteksi = True
    
if watermark_tidak_ada==True and  watermark_terdeteksi==True:
    print("Watermark Tidak Terdeteksi")
else:
    print("Watermark Terdeteksi")
```
